```python
from PIL import Image
```

```python
high_boost = Image.open('lady.jpg')
high_boost_pixels = high_boost.load()
```

```python
print("High boost before")
display(high_boost)
```

```python
for i in range(1, high_boost.size[0]-1):
    for j in range(1, high_boost.size[1]-1):
        r,g,b = high_boost_pixels[i, j]
        r += 4*high_boost_pixels[i, j][0] -high_boost_pixels[i-1, j][0] -high_boost_pixels[i+1, j][0]-high_boost_pixels[i, j-1][0]-high_boost_pixels[i, j+1][0]
        g += 4*high_boost_pixels[i, j][1] -high_boost_pixels[i-1, j][1] -high_boost_pixels[i+1, j][1]-high_boost_pixels[i, j-1][1]-high_boost_pixels[i, j+1][1]
        b += 4*high_boost_pixels[i, j][2] -high_boost_pixels[i-1, j][2] -high_boost_pixels[i+1, j][2]-high_boost_pixels[i, j-1][2]-high_boost_pixels[i, j+1][2]
        high_boost_pixels[i, j] = (r,g,b)
```

```python
print("High boost after")
display(high_boost)
```
