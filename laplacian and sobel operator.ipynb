```python
# EDGE DETECTION USING LAPLACE AND SOBEL OPERATORS
%matplotlib inline
import os
os.environ['OPENCV_IO_MAX_IMAGE_PIXELS']=str(2**64)

import cv2
import numpy as np
from matplotlib import pyplot as plt

#img = cv2.imread('img1.jpeg')
img1 = cv2.imread('img.jpeg',0)
#img1.show()

# remove noise

img = cv2.GaussianBlur(img1,(3,3),0)

# convolute with proper kernels
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y

plt.figure()
plt.imshow(img)
plt.figure()
plt.title("LAPLACIAN")
plt.imshow(laplacian) 
plt.figure()
plt.title("SOBELX")
plt.imshow(sobelx)
plt.figure()
plt.title("SOBELY")
plt.imshow(sobely)

sobel = sobelx + sobely
plt.figure()
plt.title("SOBEL")
plt.imshow(sobel)

plt.show()
```

```python
# Calculate the sharpened image
sharp = img - 0.8*laplacian

print(sharp)
plt.figure()
plt.imshow(img)
plt.figure()
plt.title("LAPLACIAN SHARPENING")
plt.imshow((sharp).astype(np.uint8))
```
