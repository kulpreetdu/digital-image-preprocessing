# AND Operator

```python
image1 = Image.open("art.jpg")
pixel1 = image.load()
```

```python
image1
```

```python
image2 = Image.new("RGB", (image1.size[0], image1.size[1]), (0, 0, 0))
pixel2 = image2.load()
for i in range(50, image2.size[0]-50):
    for j in range(50, image2.size[1]-50):
        pixel2[i,j] = (255, 255, 255)
```

```python
image2.save("mask_and.jpg")
```

```python
rows = image1.size[0]
cols = image1.size[1]
new_and_image = Image.new("RGB", (rows, cols))
pixels = new_and_image.load()
for i in range(rows):
    for j in range(cols):
        pixels[i,j] = (pixel1[i,j][0] & pixel2[i,j][0], pixel1[i,j][1] & pixel2[i,j][1], pixel1[i,j][2] & pixel2[i,j][2])
```

```python
new_and_image.save("mask_and_final.jpg")
```

# Or Operator

```python
image1 = Image.open("art.jpg")
pixel1 = image1.load()
```

```python
image2 = Image.new("RGB", (image1.size[0], image1.size[1]), (255, 255, 255))
pixel2 = image2.load()
for i in range(50, image2.size[0]-50):
    for j in range(50, image2.size[1]-50):
        pixel2[i,j] = (0, 0, 0)
```

```python
image2
```

```python
image2.save("mask_or.jpg")
```

```python
rows = image1.size[0]
cols = image1.size[1]
new_or_image = Image.new("RGB", (rows, cols))
pixels = new_or_image.load()
for i in range(rows):
    for j in range(cols):
        pixels[i,j] = (pixel1[i,j][0] | pixel2[i,j][0], pixel1[i,j][1] | pixel2[i,j][1], pixel1[i,j][2] | pixel2[i,j][2])
```

```python
new_or_image
```

```python
new_or_image.save("mask_or_final.jpg")
```
